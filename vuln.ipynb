```python
df = df.dropna(subset=['vulnType'])

# Handle missing text data in 'details' (e.g., fill with 'Unknown' or a placeholder)
df['details'].fillna('Unknown', inplace=True)

# You may also want to clean version and CVE columns if they have unwanted characters or missing data
df['versions'] = df['versions'].apply(lambda x: str(x).replace('<', 'less_than_').replace('>', 'greater_than_'))
df['CVE'] = df['CVE'].apply(lambda x: str(x) if pd.notna(x) else 'Unknown')
```

```python
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from transformers import Trainer, TrainingArguments

# Initialize the tokenizer (You can switch to CodeBERT as needed)
tokenizer = RobertaTokenizer.from_pretrained('roberta-base')

# Tokenize the 'details' column (max_length can be adjusted)
def tokenize_function(examples):
    return tokenizer(examples['details'], padding="max_length", truncation=True, max_length=128)

# Apply tokenizer to your DataFrame
df['input_ids'] = df['details'].apply(lambda x: tokenizer.encode(x, truncation=True, padding='max_length', max_length=128))
```

```python
from sklearn.model_selection import train_test_split
import torch

# Encode the labels (vulnType)
label_map = {label: idx for idx, label in enumerate(df['vulnType'].unique())}
df['labels'] = df['vulnType'].map(label_map)

# Prepare the dataset for training
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Convert to Torch Dataset for Huggingface
class VulnerabilityDataset(torch.utils.data.Dataset):
    def __init__(self, df, tokenizer):
        self.input_ids = df['input_ids'].tolist()
        self.labels = df['labels'].tolist()
        self.tokenizer = tokenizer
        
    def __len__(self):
        return len(self.input_ids)
    
    def __getitem__(self, idx):
        item = {'input_ids': torch.tensor(self.input_ids[idx], dtype=torch.long),
                'labels': torch.tensor(self.labels[idx], dtype=torch.long)}
        return item

train_dataset = VulnerabilityDataset(train_df, tokenizer)
val_dataset = VulnerabilityDataset(val_df, tokenizer)

# Initialize the model (RoBERTa for classification)
model = RobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=len(label_map))
```

```python
training_args = TrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=3,              # number of training epochs
    per_device_train_batch_size=16,  # batch size for training
    per_device_eval_batch_size=16,   # batch size for evaluation
    evaluation_strategy="epoch",     # evaluate every epoch
    logging_dir='./logs',            # directory for storing logs
)

trainer = Trainer(
    model=model,                         # the model to be trained
    args=training_args,                  # training arguments
    train_dataset=train_dataset,         # training dataset
    eval_dataset=val_dataset,            # evaluation dataset
)

# Train the model
trainer.train()
```
